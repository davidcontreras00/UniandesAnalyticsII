Split the data into training and testing sets

Train a Random Forest Regressor to predict the price of a car using the nominal features

```python
# Dummy y categoricas sobre X train y se aplican en test
# Guarda el Ã­ndice y para ECM, selecciona indices que no son de train
num_df = df.select_dtypes(include=['float64', 'int64']).copy()
num_df.head()
```

# Exercise 13.4

Apply two other methods of categorical encoding

compare the results

### Binary

```python
import category_encoders as ce # Importa libreria a usar
# define X and y
Xb = df.drop(['price'], axis=1)
yb = df.price

Xb_ = ce.BinaryEncoder().fit_transform(Xb) # BinaryEncoder() - No crea dummys sino binarios de cada categorÃ­a y luego una columna por cada bit
# ce.BinaryEncoder() - Se usa cuando se tienen muchas categorÃ­as
```

```python
Xb_.head()
```

```python
(((pd.Series(cross_val_score(RandomForestRegressor(n_estimators=10, n_jobs=-1), Xb_, yb, cv=10, scoring='neg_mean_squared_error')).describe())*-1)**(0.5))['mean']
```

### Feature Hashing

```python
# define X and y
Xh = df.drop(['price'], axis=1)
yh = df.price
```

```python
Xh_ = ce.HashingEncoder(n_components=8).fit_transform(Xh) # RepresentaciÃ³n ordinal de todas las categorÃ­as. Genera el mismo nÃºmero de variables inicial
```

```python
Xh_.head()
```

```python
(((pd.Series(cross_val_score(RandomForestRegressor(n_estimators=10, n_jobs=-1), Xh_, yh, cv=10, scoring='neg_mean_squared_error')).describe())*-1)**(0.5))['mean']
```

#### Se observa en las 3 metodologÃ­as usadas en el ejercicio, que la que alcanza mejores estimaciones es la realizada con el mÃ©todo Feature Hashing
